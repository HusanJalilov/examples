```python
import matplotlib.pyplot as plt
import math
```

```python
def show_plot(x, y):
    """
    Given x and y values, this function will plot the graph
    Args:
        x(int): fist point
        y(int): second point
    Returns:
        None
    """
    plt.scatter(x,y)
    plt.show()
    # write your code here
show_plot(4,3)
```

```python
def plot_function(k):
    """
    plot a graph of the kx function.
    Asign the x and y values to a list
    x values are from 0 to 10, y = kx
    
    Args:
        k(int): integer value
    Returns:
        None
    """
    x=[i for i in range(11)]
    y=[j*k for j in range(11)]
    print("x = ",x)
    print("y = ",y)
    plt.scatter(x,y)
    plt.show()
    # write your code here
plot_function(-2)
```

```python
plot_function(-2) 
```

```python
def plot_function_1(k):
    """
    plot a graph of the kx^2 function.
    Assign the x and y values to a list
    x values are from -10 to 11, y = kx^2
    
    Args:
        k(int): integer value
    Returns:
        None
    """
    x = [i for i in range(-10,11)]
    y = [k*j*j for j in range(-10,11)]
    plt.scatter(x,y)
    plt.show()
    # write your code here
plot_function_1(5)
```

```python
plot_function_1(5)
```

```python
def plot_sin_function():
    """
    draw a graph of the sin function
    Assign the x and y values to a list
    x values are from -500 to 501, y = sin(x)
"""
    x = []
    y = []
    for i in range(-500,501,49):

        x.append(i)
        y.append(math.sin(i))
    plt.plot(x,y)
    plt.show()
    # write your code here
plot_sin_function()
```
